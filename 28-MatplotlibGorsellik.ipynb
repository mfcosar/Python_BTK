## 28-MatplotlibGorsellik

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(ornekFigure, ornekEksen) = plt.subplots()   
# figure u çizdirmek için aynı cell içinde çağrılması gerek
ornekEksen.plot(numpyDizisi1, numpyDizisi2,color="#2DB6BB" , alpha=0.3)
ornekEksen.plot(numpyDizisi2, numpyDizisi1,"b")

# renklerin hex codelarını bularak o rengi verebiliriz. alpha değeri default =1 dir, düşürerek çzigiye transparency verilirS
```

```python
(figure2, eksen2) = plt.subplots()
eksen2.plot(numpyDizisi1, numpyDizisi1 +2, color="blue", linewidth=5.0, linestyle=":")
eksen2.plot(numpyDizisi1, numpyDizisi1 +3, color="yellow", linewidth=3.0, linestyle="-.")
eksen2.plot(numpyDizisi1, numpyDizisi1 +4, color="#BB4615", linewidth=6.0, linestyle="--")
eksen2.plot(numpyDizisi1, numpyDizisi1 +5, color="#89BB24", linewidth=6.0)
## linestyle="--" , :, -.,  gibi kullanılabilir
# marker'lar:

eksen2.plot(numpyDizisi1, numpyDizisi1 +6, color="black", linewidth=6.0, linestyle="--", marker="o", markersize=8, markerfacecolor="red")
eksen2.plot(numpyDizisi1, numpyDizisi1 +7, color="black", linewidth=6.0, linestyle="--", marker="+", markersize=5)
```
