```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Jupiter notebok'a özel bir komut, diğer editörlerde çalışmaz. Matplot grafiklerinin kolayca gösterimi içindir
%matplotlib inline 
```

```python
yasListesi= [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,70,80,75,90,80,80,85,70]
```

```python
# numpy'dan çok şaşmamak iyi olur. Bu nedenle verilen listeleri numpy dizisi olan array'lere çevirmek iyi olur
```

```python
numpyYasListesi  = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r")    #ilk verilen liste x eksenine, 2.liste y eksenine yazılır
plt.xlabel("Yaş")
plt.ylabel("Kilo")
# plt.show()
plt.title("Kilonun yaşa göre değişim grafiği")
# Bu şekilde grafik çizilir ama jupiter'dan başka bir editörde plt.show() demeden grafik çizilmez!
```

```python
# Yeni örnek yapmak için numpy hızlı dizi oluşturma fonk.dan kullanalım
```

```python
numpyDizisi1 = np.linspace(0,10,20)   # 0la 10 arasında 20 değişken oluştur, hepsi eşit aralıklı olsun
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")   # g'den sonra -- dashed demek
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g*-") 
```

```python
## 2 tane ayrı grafik yapmak: yanyana da alt alta da yazmak mümkün
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2, "r*-")
plt.subplot(1,2,2)    # 1 sıra olacak,2 column olacak, şu an 2.grafiği çiziyorum
plt.plot(numpyDizisi2, numpyDizisi1, "g*-")
```

```python
yeniFigure = plt.figure()   ## Ayni cell içinde yapmak iyi olur
figureAxes = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2, "g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik başlığı")
```

```python
#axes değerlerinin ne olduğunu anlamak için örnek:
# aynı figürü farklı eksenler tanımlayarak içiçe 2 ayrı grafikte gösterildi.

figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2, "g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1, "b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
# Yukardaki işlem subplotlar ile de yapılabilir.

(ornekFigure, ornekEksenler) = plt.subplots(nrows=1, ncols=2)   # bu çağrı 1 ekseni olan bir grafik çıkarıyor. Ama plt.figure() deseydik karşımıza 0 eksenli bir grafik çıkaracaktı
# bir tuple döndürür  ornekFigure ve eksenler döndürür
#  elimizde eksen varsa: eksen.plot() diyerek çizdirilebilir. ama birden fazl eksen olduğu için for loop'a almak gerekir
type(ornekEksenler)
#ornekEksen.plot(numpyDizisi1, numpyDizisi2, "b")

for eksen in ornekEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2, "g")
    eksen.set_xlabel("X ekseni")
    eksen.set_ylabel("Y ekseni")

## jupiter notebook dışındaki editörlerde:  ornekFigure    diye çağırmak gerekebilir.

plt.tight_layout()   # 2 subplot grafiği arasındaki mesafeyi açıp düzgün gösterim için kullanılır.
```

```python
# Yeni figure üzerinde gelişmiş özellikleri kullanalım.

yeniFigure = plt.figure()   # argument olarak figsize alabilir. Ama gerekli değil.
#  plt.figure()  : dpi verilebilir. dpi çözünürlüğü gösterir. 100'de büyük ve net; 50'de küçük ve zor seçilen bir grafik çıkar.S
## dpi=150  ==> büyük ve net olduğu için çalışması zor. En son kayıt etme anında dpi verilip çalışma bitirilebilir.

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2, label="Numpy dizisi ** 2")

#bir grafikte sadece 1 çizgi olmak zorunda değil. Başka çizgiler de çzizilebilir
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label="Numpy dizisi ** 3")
yeniEksen.legend()   # legend'e loc=1,2,3,4 ten biri verilerek gösterildiği yer değiştirilebilir: legend(loc=2)  gibi
```

```python
# Figure nasıl kaydedilir?  savefig() ile.. hangi folder'da çalışıyorsa oraya .png dosyası olarak kaydedilir
yeniFigure.savefig("ornekFigure.png", dpi=200)
```
